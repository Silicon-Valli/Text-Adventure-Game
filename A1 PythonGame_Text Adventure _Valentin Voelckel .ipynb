```python
"""Docstring:

    Title: Indiana Jones and the Last Crusade

    A) Introduction: 
    Inspired by the Indiana Jones Saga, this game lets you experience a quest
    in an undiscovered jungle.
    
    You will be sent on a quest to find the holy grail, which is located in a 
    remote temple. On the way there have to master unforeseen challenges
    and prove that you have the right gut feeling. 

    This game consists of three stages:
    
    1) The Basecamp: Put the right item into your backpack to pass obstacles. 
    2) The Path: Your destiny lies in the hands of the gods. You roll the dice to 
    decide whether to go left or right. Only one path will lead you to the temple.
    3) The Key to the Grail: The guards demand a password for you to enter. 
    Use your cheat-sheet and guess smart!
    

    B) Known Bugs and/or Errors: 
    Line: 172
    Context: The conditional statement doesn't recognize the answer 'no'. 

"""

#import packages 
import random
import time
import sys


#fail function to end game when wrong decisions have been made
def fail():
    print_pause("\nYou've failed, but thanks for playing!")
    input(prompt="<Press enter exit>\n")
    
    #restart game option
    play_again=input(prompt="Play again?? yes/no \n").lower()
    if play_again =="yes":
        basecamp()
        
    #exit    
    else:
        print("\nTill' the next adventure!")
        

#function which is called when game is completed
def success():
    print("\nYou've succeeded, thanks for playing!")
    print(" WIN " * 10)
    
    #restart game option
    play_again=input(prompt="Play again?? yes/no \n").lower()
    if play_again =="yes":
        basecamp()
        
    else:
        print("Till the next adventure!")
        
        
#make text appear slow, first define pause                 
def print_pause(text): 
    print(text)
    time.sleep(0.3)
    
    
#this is a loop to check whether input for list: items_basecamp is valid
#'text' is for user-input, 'check_list' checks items_basecamp
def valid_input(text, check_list): 
    while True:
        response = input(text).lower()
        for option in check_list:
            if option in response:
                response = option
                return response
        print_pause("Sorry, this isn't on the list")  
        
    
#introduction into the game    
def intro():  
    print("Welcome to your Indiana Jones Quest")
    time.sleep(1.6) #slows output down
    
    
#ask for adventurer name 
    print_pause("\nPlease enter your last name\n")
    user_name = input(prompt="")
    
    
#ask for location    
    print_pause(f"You find yourself in the jungle Mr. Indiana {user_name}!\n")
    print_pause("What's the name of the jungle you are going on a treasure hunt?\n")
    location_name = input(prompt="")
    
    
#start game    
    print_pause(f"Are you ready to go on an adventure in the {location_name} jungle?\n")
    input(prompt="press enter to start\n")
    basecamp()

    

##############################################################################        
#first game, name: The Basecamp   

#choose the right item for the right enemy
def basecamp():
    items_basecamp = ["whip","climbing gear"] #list of items to choose from
    enemies = ["snake", "landslide"] #enemies that can randomly appear
    decisions = ['a','b']

    
    #adding story
    print("\nStage 1:")
    print_pause("You are about to leave the basecamp and prepare yourself for a dangerous hike\n")
    print_pause("You can only select one item at a time\n")
    print_pause("\nselect item")
    
    #print items from list basecamp 
    print_pause(f"> {items_basecamp[0]}\n"
                f"> {items_basecamp[1]}\n"
               )
    
    
    #checking if input is valid by comparing it to the list
    item_backpack = valid_input("Please, enter a name of the item.\n", items_basecamp)
    print_pause(f"{item_backpack} was put in backpack\n")      
      
        
    #alarming user
    input(prompt="> press enter to continue\n")
    print_pause(f"WATCH OUT! \n"*2) 
    
    
    #enemy appears here randomly
    enemy = random.choice(enemies) #selected from list 
    print_pause("Suddenly you face an obstacle...\n")
    print_pause(f"A {enemy} appears!!\n")
    print_pause("Select a. or b.")
    
    
    #offer user options on how to react and stores them 
    use_item = print_pause(f"\na. Use your {item_backpack}")
    go_back  = print_pause("b. Go back to basecamp to choose different item")

    choice_user = valid_input("\nEnter your deciscion\n",decisions).lower() #storing input here
    
    
    #nested condition to validate input and react to scenarios
    #snake can be defended with a whip but the user has the option to return to basecamp
    #or statement checks if user input is similar to the wording in the selected option
    
    #first outcome to reach the next level
    if item_backpack == "whip" and enemy == "snake" and choice_user == 'a':
        print_pause(f"""Hmm... the snake looks like a Python! 
An animal you don't want to harm because you LOVE PYTHON\n""")
        print_pause("Use whip anyway?")
        
        #define user input
        use_anyway = input(prompt = "yes/no\n").lower()
        
        #nested condition if yes 
        if use_anyway == 'yes' or 'y':
            print_pause("You use the whip and wrap it skillfully around a limb above you\n")
            print_pause("...take off the ground and fly over the Python. NO snake got hurt!!\n\n")
            print_pause("The obstacle has successfully been passed!\n")
            path()
            
        #fail the game if answer is no    
        elif use_anyway == 'no' or 'n':
            print_pause("The snake attacks you and you die")
            fail()
            
        else:
            print("something went wrong, you go back to the camp\n")
            basecamp() 
            
            
    #second outcome to reach the next level    
    elif item_backpack== "climbing gear" and enemy == "landslide" and choice_user == 'a': 
        print_pause("Your instincts were right!\n")
        print_pause("The gear helps you to climb over the huge rocks!\n")
        print_pause("The obstacle has successfully been passed!\n")
        path()
    
    
    #Giving the player a hint that they might took a wrong decision
    elif item_backpack == "whip" and enemy == "snake" and choice_user == 'b': 
        print_pause("This tool seemed to be useful! You are going back to the camp. Next time there might be a different obstacle\n")
        basecamp()
    
    
    #Giving player a hint that they might took a wrong decision
    elif item_backpack == "climbing gear" and enemy == "landslide" and choice_user == 'a': 
        print_pause("This tool seemed to be useful! You are going back to the camp. Next time there might be a different obstacle\n")
        basecamp()    
    
    elif choice_user == 'b':
        print_pause("You run back to the camp\n")
        basecamp()   
        
    else:
        print("That didn't work! You run back to the basecamp")
        basecamp()
        
        
        
        
##############################################################################        
#second game, name: The Path 
                
#Stage two starts here 

def path():
    #storytime
    print_pause("Stage 2: Your journey continues with a hike up a mountain...")
    print_pause("...until the roads SPLIT")
    
    #interactive input to express the user's feelings
    annoyed = input(prompt="Full of anger you scream:\n")
    print_pause(annoyed *3 + "!!!\n")
    
    #intro for dice roll
    print_pause("go left or right...??\n")
    print_pause("You choose to let the destiny decide which way you will take")
    print_pause("""...Good that you find two ancient dices in your backpack from a 
previous treasure hunt!\n""")
    print_pause("You decide to go LEFT if the result is LOWER than 6")
    print_pause("...and go RIGHT if the result is 6 or HIGHER\n")
    print_pause(f"""if BOTH dices show the SAME number you roll them AGAIN 
because the gods of the jungle demand it\n""")
    
    roll = roll_dice() #trigger dice roll
    
    
#define dice roll; two dices randomized
def roll_dice():
    input(prompt="Press ENTER to roll the dice...\n")
    dice_1 = random.randint(1, 6)
    dice_2 = random.randint(1, 6)
    
    
#shows result and prints out roll 
    result = dice_1 + dice_2
    print(f"You roll {result} with {dice_1} and {dice_2}\n")
    

#if dice show the same value, throw again
    if dice_1 == dice_2:
        print("The gods demand to roll again")
        roll_dice()
        
        
#if result higher than 6, turn left        
    elif result <=6:
        print("left")
        print_pause("You end up at a Strong river\n")
        print_pause("Jump 4 times over the rocks to get to the other side\n")
        
        
#loop for jumps over rocks; built in fail()      
        presses = 3       
        while presses > 0:
            input(prompt='<Press enter>\n')
            presses -= 1   
        print_pause("You slip on the third jump and end up falling down the waterfall")
        print("You get washed up in the valley and have to hike again\n")
        path()           
            
    else:
        print("right")
        guessing() 
        
        
        
        
############################################################################## 
#third game, name: The Key to the Grail

#guessing game; guessing a single word from a list
def guessing():
    guess_list = ['untu', 'chau' , 'uxtu' , 'wini' , 'maki']
    
    #storytime
    print_pause("Looks like you took the right path!\n")
    print_pause("You arrive at the ancient temple where the holy grail is hidden!\n")
    
    input(prompt="Press enter to enter the temple\n")
    
    #explenation
    print_pause("*Suddenly movement* .... You are surrounded by a tribe guarding the temple")
    print_pause("...they request the password\n")
    print_pause("If you guess right they will let you pass but attack you after the third wrong guess\n")
    print_pause("During your research you created a list with potential passcodes for this scenario")
    print_pause("Take a look at the list- it will help!\n\n")
    print_list = input(prompt="Press enter to take a look\n")
    
    print(guess_list) #give out list of potential password
     
    
    #random word that has to be guessed is generated here
    password = random.choice(guess_list) 
    print_pause("Guess the characters or word") 
    
    #creating empty list to store turns
    guesses = '' 
  
    # defining the number of turns
    turns = 4
  
    #loop to check if input matches random password 
    while turns > 0: 
      
        # counts the failed attempts
        failed = 0
      
        #check characters with list
        for letter in password:  
            if letter in guesses:  
                print(letter) 
            
            #if letter not guessed, placeholder for missing charakters is being printed
            else:  
                print("*") 
              
                #adding failed guess
                failed += 1
                
              
        #checking if player failed
        #if user didn't guess wrong he wins and it's given as an output
        if failed == 0: 
            print("\nYou guessed right")  
          
            # this print the correct word 
            print("\nThe password is: ", password)  
            success()
            break
    
      
        #input for guess
        guess = input(prompt="guess a character or word:") 
      
        #guessed letter are stored in guesses and counted
        guesses += guess  
      
        #letters from input are being checked with characters from password
        if guess not in password: 
          
            turns -= 1
          
            #if letter doesn't match 
            print("Wrong") 
          
            # number of turns are printed
            print("You have", + turns, 'more guesses') 
          
            # if all turns are used player looses
            if turns == 0: 
                print("You Lost") 
                fail()
                
#initiate game                
intro()
```
